```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('students_score.csv')
```

```python
print(df.head())
```

```python
df.isnull().sum()
```

Drop Unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
```

Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

From the above the chart we can analysis that the number of female students is more than the number of male students

```python
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
```

```python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
print(gb1)
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(gb1,annot=True)
plt.show()
```

from above plot we have concluded that there is no/negligible impact on the students score due to their parents marital status.
